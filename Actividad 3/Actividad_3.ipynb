```python
class Numero():
    def __init__(self, number):
        self.normal = number
        self.romano = self.toRoman(number)

    def toRoman(self, number):
        numeros = [
            (1000, "M"), (900, "CM"), (500, "D"), (400, "CD"), 
            (100, "C"), (90, "XC"), (50, "L"), (40, "XL"),
            (10, "X"), (9, "IX"), (5, "V"), (4, "IV"), (1, "I")
        ]
        
        resultado = ""

        for valor, simbolo in numeros:
            while number >= valor:
                resultado += simbolo
                number -= valor

        return resultado

    def toNumber(self, romano):
        caracteres = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
        total = 0
        prev = 0

        for char in reversed(romano):
            actual = caracteres[char]
            if (actual < prev):
                total -= actual
            else:
                total += actual
            prev = actual
        
        return total

    def suma_romano(self, romano):
        if not self.is_romano(romano):
            raise Exception()
        number = self.toNumber(romano)

        self.normal += number
        self.romano = self.toRoman(self.normal) 

    def imprime(self):
        print(f"{self.normal} {self.romano}")

    def is_romano(self, string):
        caracteres = ['I', 'V', 'X', 'L', 'C', 'D', 'M']

        for char in string:
            if char not in caracteres:
                return False
        
        return True
```
